```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import cv2
```

```python
outpath = './output_files/'
if not os.path.exists(outpath):
    os.makedirs(outpath)
plt.rcdefaults()
```

```python
filepath = os.path.join("./input_files/")
files = {
    "A" : "cells_positions_CAF1.csv",
    "B" : "cells_positions_CAF2.csv",
    "C" : "cells_positions_immune.csv",
}

immune_df = pd.read_csv(os.path.join(filepath, files["C"]))
caf1_df = pd.read_csv(os.path.join(filepath, files["A"]))
caf2_df = pd.read_csv(os.path.join(filepath, files["B"]))
display(immune_df, caf1_df, caf2_df)
```

```python
# from sklearn.neighbors import RadiusNeighborsClassifier

def brute_find_rnn(query_pos, search_pos, verbose = True):
    """
    brute force find radius nearest neighbours
    Note: scipy's kDtree would work but slightly less accurate.
    sklearn's version does not accept query and search points (only 1 set of points)
    """
    
    nbrs_per_x = []
    
    for n in range(query_pos.shape[0]):
        
        x = query_pos[[n],:]
        if verbose: print("Query point:", x, x.shape)
        
        diff = search_pos - x
        if verbose: print(diff.shape)
        
        dist = np.linalg.norm(diff, axis=1)
        if verbose: print(dist.shape)
        
        radius_nbrs_mask = (dist < max_r_pix)
        num_nbrs = np.sum(radius_nbrs_mask)
        if verbose: print(f"{num_nbrs} for cell {n}")
        
        nbrs_per_x.append(num_nbrs)
        
    return np.array(nbrs_per_x)


max_r = 100 # in um
res = 0.12 # um per pixel
max_r_pix = max_r / res

print(f"radius of {max_r_pix} pixels.")

immune_pos = immune_df.loc[:,["x","y"]].values
caf1_pos = caf1_df.loc[:,["x","y"]].values
caf2_pos = caf2_df.loc[:,["x","y"]].values
print(immune_pos.shape, caf1_pos.shape, caf2_pos.shape,)

caf1_nn = brute_find_rnn(immune_pos, caf1_pos, verbose = False)
caf2_nn = brute_find_rnn(immune_pos, caf2_pos, verbose = False)

plt.boxplot([caf1_nn, caf2_nn])
plt.xticks([1, 2], ['CAF1', 'CAF2',])
```

```python
fig, ax = plt.subplots(figsize=(6,6))
sns.boxplot(data = [caf1_nn, caf2_nn], ax=ax)
ax.set_xticks([0, 1], ['CAF1', 'CAF2',])
ax.set_ylabel(f"num cells within {max_r}um")
fig.savefig(os.path.join(outpath,"numcells_compare.jpg"))
```
